```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.patches import Rectangle
import seaborn as sns
import geopandas as gpd
```

```python
temp_prof = pd.read_pickle('./data/vre/t2m_hourly_desert_2020.pkl') #this file is to large, can download it from era5, or contact the lead contact

data = np.array(list(temp_prof.values()))
```

```python
data_valid_cell = {}

for k in temp_prof:
    if np.mean(temp_prof[k])<=25+273.15 and np.max(temp_prof[k])<=46+273.15:
        data_valid_cell[k] = temp_prof[k]
        
data_valid_cell = np.array(list(data_valid_cell.values()))
```

```python
dc_temp = pd.read_pickle('./data/vre/t2m_hourly_dc_2020.pkl')

data_dc = np.array(list(dc_temp.values()))
```

```python
data = data - 273.15
data_dc = data_dc - 273.15
data_valid_cell = data_valid_cell - 273.15
```

```python

median_ = np.median(data, axis=0)
mean_ = np.mean(data, axis=0)
q_25 = np.percentile(data, 25, axis=0)
q_75 = np.percentile(data, 75, axis=0)
q_0 = np.percentile(data, 0, axis=0)
q_100 = np.percentile(data, 100, axis=0)


median_dc = np.median(data_dc, axis=0)
mean_dc = np.mean(data_dc, axis=0)
q_25_dc = np.percentile(data_dc, 25, axis=0)
q_75_dc = np.percentile(data_dc, 75, axis=0)
q_0_dc = np.percentile(data_dc, 0, axis=0)
q_100_dc = np.percentile(data_dc, 100, axis=0)

median_vc = np.median(data_valid_cell, axis=0)
mean_vc = np.mean(data_valid_cell, axis=0)
q_25_vc = np.percentile(data_valid_cell, 25, axis=0)
q_75_vc = np.percentile(data_valid_cell, 75, axis=0)
q_0_vc = np.percentile(data_valid_cell, 0, axis=0)
q_100_vc = np.percentile(data_valid_cell, 100, axis=0)
```

```python
data.shape
```

```python

daily_range = np.ptp(data_valid_cell.reshape(data_valid_cell.shape[0], 366, 24), axis=2) 

daily_range_edc = np.ptp(data_dc.reshape(data_dc.shape[0], 366, 24), axis=2) 
```

```python

median_edc_daily = np.median(daily_range_edc, axis=0)
mean_edc_daily = np.mean(daily_range_edc, axis=0)
q_25_edc_daily = np.percentile(daily_range_edc, 25, axis=0)
q_75_edc_daily = np.percentile(daily_range_edc, 75, axis=0)
q_0_edc_daily = np.percentile(daily_range_edc, 0, axis=0)
q_100_edc_daily = np.percentile(daily_range_edc, 100, axis=0)

median_ddc_daily = np.median(daily_range, axis=0)
mean_ddc_daily = np.mean(daily_range, axis=0)
q_25_ddc_daily = np.percentile(daily_range, 25, axis=0)
q_75_ddc_daily = np.percentile(daily_range, 75, axis=0)
q_0_ddc_daily = np.percentile(daily_range, 0, axis=0)
q_100_ddc_daily = np.percentile(daily_range, 100, axis=0)
```

```python
global_mean = np.mean(mean_)
```

```python
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 25

days = range(8784)

fig,ax = plt.subplots(figsize=(14,10),dpi=600)

# 颜色映射
colors = '#7855e5'
vc_color = '#F06292'


ax.fill_between(days, q_25_vc, q_75_vc, color=colors, alpha=0.3)
ax.fill_between(days, q_0_vc, q_100_vc, color=colors, alpha=0.15)
ax.plot(days, median_vc, color=colors, linewidth=1.5)

ax.axhline(0,color='black',lw=1.4,ls=':')


ax.set_ylabel('Ambient temperature (°C)',font={"size":25})

ax.set_xticks(range(0,8760,360),range(0,8760,360),rotation=90)


ax.set_xlabel('Hours')

ax.minorticks_on()

ax.tick_params(axis='y', 
            which='major', 
            length=6, 
            color='black')

ax.tick_params(axis='y', 
            which='minor', 
            length=3, 
            color='black')

ax.tick_params(axis='x', 
            which='minor', 
            length=0, 
            color='blue')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_ylim(-60,60)
ax.set_xlim(0,8760)

legend_ax = fig.add_axes([0.55, 0.125, 0.05, 0.15])

legend_ax.axis('off')

legend_ax.spines['top'].set_visible(False)
legend_ax.spines['right'].set_visible(False)
legend_ax.spines['bottom'].set_visible(False)
legend_ax.spines['left'].set_visible(False)

legend_ax.set_xticks([],[])
legend_ax.set_yticks([],[])

# 定义图例参数
sources = ['MC']
x_positions = [0]

box_width = 0.5
box_height = 0.8

bottom = 0.1

# 绘制图例
for i, (source, x_pos) in enumerate(zip(sources, x_positions)):
    # 绘制100%框
    legend_ax.add_patch(Rectangle((x_pos, bottom), 
                                  box_width, 
                                  box_height,
                                  facecolor=colors,
                                  edgecolor='black', 
                                  alpha=0.15))
    
    legend_ax.add_patch(Rectangle((x_pos, bottom), 
                                  box_width, 
                                  box_height,
                                  facecolor='none',
                                  edgecolor='black'))
    
    # 绘制50%框
    legend_ax.add_patch(Rectangle((x_pos, bottom+0.25*box_height), 
                                  box_width, 
                                  box_height/2,
                                  facecolor=colors, 
                                  alpha=0.3))
    
    # 绘制中位数线
    legend_ax.plot([x_pos, x_pos + box_width], 
                   [bottom+0.5*box_height, 
                    bottom+0.5*box_height], 
                    color=colors, 
                    linewidth=1.5)
    
    if source == sources[-1]:
        legend_ax.text(x_pos + box_width + 0.025, bottom+0.5*box_height, 'Median', va='center',fontsize=18)
        legend_ax.text(x_pos + box_width + 0.025, bottom+box_height, '100%', va='center',fontsize=18)
        legend_ax.text(x_pos + box_width + 0.025, bottom+0.75*box_height, '75%', va='center',fontsize=18)

# 设置图例坐标轴的范围
legend_ax.set_xlim(-0.1, 1.1)
legend_ax.set_ylim(0.0, 1.0)


plt.savefig('./data/fig/Vis/temp_prof_vc.jpg',bbox_inches='tight')

plt.close()
```
