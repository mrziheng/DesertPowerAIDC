```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
from Visual import MapViser
import seaborn as sns
```

```python
map_vis = MapViser()
```

```python
world_land = gpd.read_file('./data/geo/Base/world_land.shp')
desert = gpd.read_file('./data/geo/desert.shp')
```

```python
desert['idx'] = 100 * range(desert.shape[0])
```

```python

map_vis.set_font()

map_vis.set_bound_color('gainsboro')

fig,ax = map_vis.draw_attr_map(vis_shp=desert,
                                attr_col='idx',
                                cmap='Set3',
                                out_fig='./data/fig/Vis/pv_gen_poten.png',
                                cbar_label='Generation potential (TWh/yr)',
                                #ub=0.25,
                                with_cbar=False,
                                bound_mode=2,
                                return_fig=True)

ax.set_xlim(-180,180)


plt.savefig('./data/fig/Vis/desert_area.jpg',bbox_inches='tight')
```
