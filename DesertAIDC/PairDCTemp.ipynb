```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
from Visual import MapViser
import seaborn as sns
```

```python
data_center = gpd.read_file('./data/geo/dc_loc.gpkg')
temp = pd.read_csv('./data/vre/t2m_global_2020.csv')
```

```python
data_center['temp_mean'] = -1.0
data_center['temp_max'] = -1.0
data_center['temp_min'] = -1.0
```

```python
temp.loc[temp['lon']>180,'lon'] = temp.loc[temp['lon']>180,'lon'] - 360
```

```python
temp = temp.set_index(['lon','lat'])
```

```python
for i,r in data_center.iterrows():
    lon = r.geometry.x
    lat = r.geometry.y
    
    lon_x = (lon//0.25) * 0.25
    lat_x = (lat//0.25+1) * 0.25
    
    
    temp_mean = temp.at[(lon_x,lat_x),'mean']
    temp_max = temp.at[(lon_x,lat_x),'max']
    temp_min = temp.at[(lon_x,lat_x),'min']
                
    data_center.at[i,'temp_mean'] = temp_mean
    data_center.at[i,'temp_max'] = temp_max
    data_center.at[i,'temp_min'] = temp_min
```

```python
sns.reset_defaults()

plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 25
# 
fig,ax = plt.subplots(figsize=(18,12),dpi=600)


for x,t in enumerate(['temp_min','temp_mean','temp_max']):
    sns.violinplot(ax=ax,
                   x=x, 
                   y=t, 
                   data=data_center)
    

ax.grid(which='both')

ax.set_ylim(-50,50)
    
#ax.set_yticks(range(-50,51,10),range(-50,51,10),font={'size':20})

ax.set_xticks([0,1,2],['Min','Mean','Max'],font={'size':25})

ax.tick_params(axis='y', 
            which='major', 
            length=8, 
            color='black')

ax.tick_params(axis='x', 
            which='major', 
            length=8, 
            color='black')
    
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


ax.set_ylabel('Ambient temperature (C)',font={"size":25})

plt.savefig('./data/fig/Vis/temp_violin.jpg',bbox_inches='tight')

# 
plt.show()
```
